```python
import requests
```

```python
json_url = "https://www.bikeseoul.com/app/station/getStationRealtimeStatus.do"
```

```python
res = requests.get(json_url)
res
```

```python
# json 문자열을 => 파이썬 객체화

import json

response_dict = json.loads(res.text)
response_dict.keys()
```

```python
station_list = response_dict["realtimeList"]
len(station_list)
```

```python
import pandas as pd

df = pd.DataFrame(station_list)
print(df.shape)
df.head()
```

# 지도 시각화

```python
import folium

center = [45.5236, -122.6750]

m = folium.Map(location=center, zoom_start=13)

marker = folium.Marker(
    center,
    popup="팝업 메시지",
    tooltip="툴팁 메시지",
)
marker.add_to(m)

m
```

```python
center_lat = sum([float(station["stationLatitude"]) for station in station_list]) / len(station_list)
center_lng = sum([float(station["stationLongitude"]) for station in station_list]) / len(station_list)
center_lat, center_lng
```

```python
center_lat = df["stationLatitude"].astype("float").mean()
center_lng = df["stationLongitude"].astype("float").mean()
center_lat, center_lng
```

```python
folium_map = folium.Map(
    [center_lat, center_lng],
    zoom_start=15,
)
folium_map
```

```python
station_list[0]
```

```python
folium_map = folium.Map(
    [center_lat, center_lng],
    zoom_start=15,
)

for station in station_list:
    name = station["stationName"]
    lat = station["stationLatitude"]
    lng = station["stationLongitude"]
    marker = folium.Marker(
        [lat, lng],
        tooltip=name,
    )
    marker.add_to(folium_map)

folium_map
```
