```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://m.comic.naver.com/webtoon/list?titleId=730656&page=1&week=tue&sortOrder="

res = requests.get(url)
res
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
}
```

```python
html = res.text
html
```

```python
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
title_list = soup.select('span.name strong') 
len(title_list)
```

```python
item_tags = soup.select('li.item')

url_list1 = []

for item_tag in item_tags:
    title_tag = item_tag.select_one('span.name strong')
    if title_tag:
        title = title_tag.get_text()
        print(title)
    url_tag = item_tag.select_one('a')  
    if url_tag:
        url = url_tag['href']
        url_list1.append(url)             
        print(url)
```

```python
 # 위 리스트에 불필요한 정보들은 빼고 필요한 것만 추출

url_list2 = []       
for i in range(len(url_list1)):   
    if url_list1[i].startswith("/webtoon/detail"):
        url_list2.append(url_list1[i])
        
url_list2
```
