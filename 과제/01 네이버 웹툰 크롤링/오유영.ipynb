```python
import requests
from bs4 import BeautifulSoup
```

```python
# 1. 로그인없이 볼 수 있는 웹툰 하나 선택

url = "https://m.comic.naver.com/webtoon/list?titleId=796152&page=3&sort=DESC"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
}
```

```python
res = requests.get(url, headers=headers)
res
```

```python
html = res.text
len(html)
```

```python
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
# 2. 그 웹툰의 첫번째 페이지를 크롤링해서 회차와 그 회차의 URL을 리스트로 뽑아낸다.

w_list = soup.select(".section_episode_list a")
w_list
```

```python
w_list = soup.select(".section_episode_list a")
len(w_list)
```

```python
#이미지 뽑아내기

w_image = soup.select(".thumbnail img")
len(w_image)
```

```python
# 3. 크롤링한 모든 회차의 이미지를 웹툰명/회차/이미지파일명.jpg 경로에 모두 저장

w = w_image[0]
w
```

```python
w["src"]
```

```python
for idx, w in enumerate(w_image, 1):
    filename = f"{idx}.jpg"
    print(filename)
    
    image_res = requests.get(w["src"], headers=headers)
    assert image_res.status_code == 200, "서버 응답이 200이 아니예요!!!"
    
    #저장
    with open(filename, "wb") as f:
        f.write(image_res.content)
```

```python
# 3. 크롤링한 모든 회차의 이미지를 웹툰명/회차/이미지파일명.jpg 경로에 모두 저장

import os
from tqdm.notebook import tqdm

웹툰명 = "마루는강쥐"
회차 = "1-29화"

for idx, w in tqdm(enumerate(w_list, 1)):
    filepath = f"{웹툰명}/{회차}/{idx:03d}.jpg"
    print(filepath)
    
    dirpath = os.path.dirname(filepath)
    os.makedirs(dirpath, exist_ok = True)
    
    # w.select("img")[0]
    img_url = w.select_one("img")["src"]
    
    print(img_url)
    image_res = requests.get(img_url, headers=headers)
    assert image_res.status_code == 200, "서버 응답이 200이 아니예요!!!"
    
    #저장
    with open(filepath, "wb") as f:
        f.write(image_res.content)
```
