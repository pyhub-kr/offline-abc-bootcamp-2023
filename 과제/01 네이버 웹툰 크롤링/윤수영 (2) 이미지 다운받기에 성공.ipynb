```python
import requests
```

```python
url = "https://m.comic.naver.com/webtoon/list?titleId=783540"
```

```python
res = requests.get(url)
res
```

```python
html = res.text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
tag_list = soup.select(".section_episode_list a")
```

```python
tag_list[4]
```

```python
import os
from urllib.parse import urljoin

headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36",
} # 네이버를 속여서 접속

웹툰명 = "여름여자 하보이"

for tag in tag_list:
    ep_url = tag["href"]
    
    if ep_url != "#":
        
        name_tag = tag.select(".name")[0]
        name = name_tag.text
        print(name)
        # print(ep_url)
        ep_url = "https://comic.naver.com" + ep_url
        ep_url = urljoin(url, ep_url)
        print(ep_url)
        
        page_url = ep_url
        res_2 = requests.get(page_url, headers = headers)
        html_2 = res_2.text
        soup_2 = BeautifulSoup(html_2, "html.parser")
        image_list = soup_2.select(".wt_viewer img")
        
        for idx, image in enumerate(image_list,1):
            
            filepath = f"{웹툰명}/{name}/{idx:03d}.jpg"
            
            dirpath = os.path.dirname(filepath) # 경로 계산하고, 없으면 생성
            os.makedirs(dirpath, exist_ok = True)

            image_res = requests.get(image["src"], headers = headers) # image["src"]는 이미지 url 주소
            assert image_res.status_code == 200, "서버 응답이 200이 아니에요!" # assert 뒤에 코드는 참이어야 된다, 만약 거짓이면 뒤에 메세지가 출력
            print(image["src"])
            
            # 저장
            with open(filepath, "wb") as f:
                f.write(image_res.content)      
```
