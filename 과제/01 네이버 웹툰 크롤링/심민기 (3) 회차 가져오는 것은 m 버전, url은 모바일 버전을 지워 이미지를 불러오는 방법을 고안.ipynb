```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin
```

```python
index_page_url = "https://m.comic.naver.com/webtoon/list?titleId=783540"
```

```python
headers = {
 "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36",
}
```

```python
res = requests.get(index_page_url)
res
```

```python
html = res.text
html
```

```python
soup = BeautifulSoup(html, "html.parser")

tag_list = soup.select(".section_episode_list a")
len(tag_list)
```

```python
for a_tag in tag_list:
    ep_url = a_tag["href"]
    
    if ep_url != "#":
        
        
        
        name_tag = a_tag.select(".name")[0]
        name = name_tag.text
        print(name)
        #print(ep_url)
        
        ep_url = "https://comic.naver.com" + ep_url
        ep_url = urljoin(index_page_url, ep_url)
        
        print(ep_url)
        
        page_url = ep_url
            
        res_2 = requests.get(page_url)
        html_2 = res_2.text
        soup_2 = BeautifulSoup(html_2, "html.parser")
        tag_list_2 = soup_2.select(".wt_viewer img")
        
        for idx, tag in enumerate(tag_list_2, 1):
            
            웹툰명 = "여름여자 하보이"
            회차 = name
            
            filename = f"{웹툰명}/{회차}/ {idx}.jpg"
            print (filename)

            filepath = filename
            dir_path = os.path.dirname(filepath) # os.path 경로 관련 기능 제공

            if not os.path.exists(dir_path):
                os.makedirs(dir_path)

            image_res = requests.get(tag["src"], headers = headers)
            assert image_res.status_code == 200, "서버 응답이 200이 아니에요!"

            print(tag["src"])

            # 이미지 저장
            with open(filename, "wb") as f:
                f.write(image_res.content)
            
        
      
        
    
```
