```python
# 라이브러리 임포트
import requests 
from bs4 import BeautifulSoup
import os
import time
!pip install tqdm
from tqdm.notebook import tqdm
```

```python
#url 저장
url = "https://m.comic.naver.com/webtoon/list?titleId=783050"
```

```python
#header 에 user agent를 저장하여 naver의 차단을 피한다.
headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
}
```

```python
#res에 url을 불러온다.
res = requests.get(url, headers=headers)
res
```

```python
#html에 text로 저장하고 tag_list에 .preciw a에 해당하는 것만 뽑아준다.
html = res.text
soup = BeautifulSoup(html, "html.parser")
tag_list = soup.select(".preview a")
```

```python
#뽑아낸 url 은 앞에 "https://m.comic.naver.com/" 가 붙어야 제대로 작동함 때문에 이따 붙여지구 위해 저장 
# 각 완성된 url들을 빈 리스트에 저장해둔다
f_tag = "https://m.comic.naver.com/"
ep_tag_list =[]
```

```python
#신군 웹툰 각 회차에 접속하는 url을 반복문을 통해 ep_tag[]에 하나씩 저장함 (출력하여 잘 저장되는지 확인)
for a_tag in tag_list:
    #if ep_url !="#"
    ep_tag = f_tag +a_tag["href"]
    tag_name = a_tag.select(".name")
    print(tag_name)
    print(ep_tag)
    ep_tag_list.append(ep_tag)
```

```python
# i_tag_list를 추출하는 함수 제작 
def img_extract(i_page_url):
    res = requests.get(i_page_url)
    html = res.text
    soup = BeautifulSoup(html, "html.parser")
    i_tag_list = soup.select("img")
    return i_tag_list
```

```python
# 각 회차(링크) 별 이미지 주소 잘 출력되는지 확인 이때 웹툰 내용과 관련없는 이미지와 출력을 방해하는 다른 tag들을 막기 위해 
#        if  i_tag["class"] ==['swiper-lazy']: 조건을 주어 방해하는 웹툰 이미지 클래스가 아닌 class가 나오면 탈출하도록 함
for i_page_url in ep_tag_list:
    i_tag_list=img_extract(i_page_url)
    
    for i_tag in i_tag_list:  
        if  i_tag["class"] ==['swiper-lazy']:
            break
        print(i_tag["data-src"])
    
```

```python
# 각 회차(링크) 별 이미지 주소가 담긴 ep_tag_list를 for중첩 반복문을 이용해서 각 링크별로 이미지를 추출해줌 

웹툰명 ="신군"
num = 80
회차 = f"{num}화"
for i_page_url in ep_tag_list:
    i_tag_list=img_extract(i_page_url)
    
    for idx, i_tag in tqdm(enumerate(i_tag_list, 1)):
    
        if  i_tag["class"] ==['swiper-lazy']:
            break    
    
        filepath = f"{웹툰명}/{회차}/{idx:03d}.jpg"
        print(filepath)
    
        dirpath= os.path.dirname(filepath)
        os.makedirs(dirpath, exist_ok=True)
    
        image_res=requests.get(i_tag["data-src"], headers=headers)
        assert image_res.status_code == 200, "서버 응답이 200이 아니예요!!"
    
        with open (filepath,"wb") as f:
            f.write(image_res.content)
    num = num-1
    회차 = f"{num}화"
```
