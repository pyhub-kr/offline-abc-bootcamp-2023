```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import os

index_page_url = "https://m.comic.naver.com/webtoon/list?titleId=777767"

웹툰명 = "역대급영지설계사"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
}

res = requests.get(index_page_url, headers = headers) # user-agent로 유저정보 확인하기 때문에, 정보를 숨겨 재요청
res
```

```python
html = res.text
len(html) # 글자수
```

```python
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
tag_list = soup.select(".section_episode_list a") # .select는 태그 리스트를 반환
len(tag_list)
```

```python
tag_list
```

```python
for a_tag in tag_list:
    ep_url = a_tag["href"]
    if ep_url != "#":
        # 태그는 다시 .select를 통해 자식 태그 조회 가능
        name_tag = a_tag.select(".name")[0]
        name = name_tag.text
        print(name)
#         print(ep_url)
        ep_url = "https://m.comic.naver.com" + ep_url
         
        # 파일 경로 설정 및 저장
        from tqdm.notebook import tqdm 

        회차 = name

        for idx, tag in tqdm(enumerate(tag_list, 1)):
            filepath = f"{웹툰명}/{회차}/{idx:03d}.jpg" # 파일 최종 경로
            print(filepath)

            dir_path = os.path.dirname(filepath) # 부모 경로 추출
            os.makedirs(dir_path, exist_ok = True) # 없으면 생성
            

            image_res = requests.get(tag["src"], headers = headers)
            assert image_res.status_code == 200, "서버 응답이 200이 아니에요!!"


            # 저장
            with open(filepath, "wb") as f:
                f.write(image_res.content)

        print(ep_url)
```
