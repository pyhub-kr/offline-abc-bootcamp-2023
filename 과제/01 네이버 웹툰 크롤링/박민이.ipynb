# 교수님 코드 기반으로 로직 짜기

```python
tag_list = soup.select(".section_episode_list a")
tag_list
```

```python
from urllib.parse import urljoin
```

```python
ep_web_url_list = []
```

```python
# 모바일 페이지 URL을 담을 리스트
ep_mobile_url_list = []

# 웹 페이지 URL을 담을 리스트
ep_web_url_list = []

for a_tag in tag_list:
    ep_url = a_tag["href"]
    if ep_url != "#":
        name_tag = a_tag.select(".name")[0]
        name = name_tag.text
        print(name)

        # 모바일 페이지 URL 설정
        ep_mobile_url = "https://m.comic.naver.com" + ep_url
        ep_mobile_url_list.append(ep_mobile_url)

        # 웹 페이지 URL 설정
        ep_web_url = "https://comic.naver.com" + ep_url
        ep_web_url_list.append(ep_web_url)

        print(ep_mobile_url)
        print(ep_web_url)
        print()
```

# 웹툰명/회차/이미지 파일명
## 교수님 웹페이지로 이미지는 다운받았지만.. 
## 파일에 똑같은 이미지만 다운이 됩니다.............

```python
for ep_idx, ep_url in enumerate(ep_web_url_list, 1):
    img_res = requests.get(ep_url, headers=headers)
    img_html = img_res.text
    
    img_soup = BeautifulSoup(img_html, "html.parser")
    img_tag_list = img_soup.select("img")
    ep_folder = f"물위의우리/{ep_idx:02d}회차"
    os.makedirs(ep_folder, exist_ok=True)  # 회차 폴더 생성
    
    for idx, image_tag in tqdm(enumerate(img_tag_list, 1), total=len(img_tag_list), desc=f"에피소드 {ep_idx:02d} 다운로드"):
        image_url = image_tag["src"]
        image_extension = image_url.rsplit(".", 1)[-1].lower()
        res = requests.get(image_url, headers=headers)
        
        try:
            res.raise_for_status()  # 오류가 발생하면 예외를 일으킴
            
            image_data = res.content
            filename = f"{ep_folder}/{idx:02d}.{image_extension}"
            
            with open(filename, "wb") as f:
                f.write(image_data)
        except Exception as e:
            print(f"이미지 다운로드에 실패했습니다. URL: {image_url}")
            print(f"에러 메시지: {str(e)}")
```
