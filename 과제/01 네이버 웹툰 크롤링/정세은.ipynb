```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://m.comic.naver.com/webtoon/list?titleId=183559"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
}
```

```python
res = requests.get(url, headers=headers)
res
```

```python
res.text
```

```python
html = res.text
len(html)
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
soup.select(".section_episode_list a")
tag_list = soup.select(".section_episode_list a")
len(tag_list)
```

```python
for tag in tag_list:
    link = tag["href"]
    label = tag.text.strip()[0:7]
    if label:
        if link != "#":
            print(repr(label))
            print(link)
            print()
```

```python
link_list = []
for tag in tag_list:
    link = tag["href"]
    if link != "#":
        link_list.append(link)
        
for i in range(0, len(link_list)):
    link_list[i] = "https://m.comic.naver.com/"+link_list[i]
    
print(link_list)
```

```python
import os

for i in range(0, len(link_list)):
    page_url = link_list[i]
    res = requests.get(page_url)
    res
    html = res.text
    soup = BeautifulSoup(html, "html.parser")
    soup.select(".wt_viewer img")
    tag_list = soup.select(".wt_viewer img")
    웹툰명 = "신의탑"
    회차 = 156-i

    for idx, tag in enumerate(link_list, 1):
        filepath = f"{웹툰명}/{회차}/{idx:03d}.jpg"
        print(filepath)
    
        dirpath = os.path.dirname(filepath) # 경로 계산하고 없으면 생성
        os.makedirs(dirpath, exist_ok = True)
    
        image_res = requests.get(tag["src"], headers=headers)
        assert image_res.status_code ==200, "서버 응답이 200이 아니예요!"
    
    # 저장
        with open(filepath, "wb") as f:
            f.write(image_res.content)
```
