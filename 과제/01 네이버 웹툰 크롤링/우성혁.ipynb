```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://m.comic.naver.com/webtoon/list?titleId=739115&week=mon&sortOrder=ASC"   #웹툰 회차 페이지 url 설정
```

```python
headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
}
# 나의 헤더 설정
```

```python
res = requests.get(url, headers=headers)
res   #응답 확인     
```

```python
res.content #응답 결과 확인
```

```python
html = res.text  #문자열이기 때문에 text로 확인 가능
len(html) #글자수 확인
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "html.parser")   #BeautifulSoup 끌어오기
```

```python
tag_list = soup.select('.section_episode_list a')
tag_list     #웹사이트에서 각 회차의 url을 크롤링
```

```python
tag = tag_list[0]
tag   #첫 번째 항목 확인
```

```python
tag["href"] #항목에서 url부분 찾기
```

```python
tag.text  #url과 매칭되는 이름 찾기
```

```python
print(tag.text.split('\n')[6])  #이름 전처리
```

```python
for tag in tag_list:
    print(tag.text.split('\n')[6])
    print(tag["href"])

    print()
```

```python
link_dict = {}

for tag in tag_list:
    link_dict[tag.text.split('\n')[6]] = tag["href"]      
    
link_dict   #딕셔너리 생성 후 회차 이름과 해당하는 url 입력
```

```python
for key in link_dict:
    print('https://m.comic.naver.com' + link_dict[key])
```

```python
import os

for key in link_dict:
    page_url = 'https://m.comic.naver.com' + link_dict[key]
    res = requests.get(page_url)
    html = res.text
    
    soup = BeautifulSoup(html, "html.parser")
    
    tag_list = soup.select(".toon_view_lst p img")

    webtoon = "앵무살수"
    series = key

    for idx, tag in enumerate(tag_list,1):    
        filepath = f"{webtoon}/{series}/{idx:02d}.jpg"      #:03d  => 3칸 확보, 3칸안에서 0붙혀서 숫자 표현
        print(filepath)

        dirpath = os.path.dirname(filepath)  #경로 계산하고, 없으면 생성
        os.makedirs(dirpath, exist_ok = True)

        image_res = requests.get(tag["src"], headers =headers)
        assert image_res.status_code == 200, "서버 응답이 200이 아닙니다"

        #저장
        with open(filepath, "wb") as f:
            f.write(image_res.content)
```
