```python
# 라이브러리 임포트
import requests
from bs4 import BeautifulSoup
from tqdm.notebook import tqdm  # tqdm로 진행률 표시
import os # 경로
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Whale/2.8.108.21 Safari/537.36",
}
```

```python
# url이 제대로 불러와지는지 상태코드 200 확인
index_page_url = "https://m.comic.naver.com/webtoon/list?titleId=761722&week=fri"
res = requests.get(index_page_url)
res
```

```python
html = res.text
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
tag_list = soup.select(".section_episode_list a")
```

```python
tag_list[0]
```

```python
웹툰명 = "세기말 풋사과 보습학원"
http = "https://comic.naver.com" # url에서 잘린 앞부분 변수로 지정
ep_list = []

for a_tag in tag_list :
    ep_url = a_tag["href"] # a_tag에서 속성이 href인 값 추출
    if ep_url != "#" : # href 속성이 #이 아닌 값(유료가 아닌 화만 추출)
        name_tag = a_tag.select(".name")[0] # a_tag에서 클래스명이 name인 값 추출
        name = name_tag.text
        print(name)
        string_list = [http, ep_url]
        result = "".join(string_list)
        ep_list.append(result)

        for i, page_url in enumerate(ep_list):
            res = requests.get(page_url)
            html = res.text
            soup = BeautifulSoup(html, "html.parser")
            tag_list = soup.select(".wt_viewer img")
            tag = tag_list[i]

            for idx, tag in tqdm(enumerate(tag_list, 1)):
                image_url = tag["src"]
                image_filepath = f"{웹툰명}/{name}/{idx:03d}.jpg"
                image_res = requests.get(image_url, headers=headers)

                dirpath = os.path.dirname(image_filepath)  # 경로 계산하고, 없으면 생성
                os.makedirs(dirpath, exist_ok=True)

                # 저장
                with open(image_filepath, "wb") as f:
                    f.write(image_res.content)
```
