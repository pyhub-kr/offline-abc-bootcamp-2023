```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Whale/2.8.108.21 Safari/537.36",
}
```

```python
url = "https://www.melon.com/chart/index.htm"
res = requests.get(url,headers = headers)
res
```

```python
html = res.text
len(html)
```

```python
soup = BeautifulSoup(html, "html.parser")
song_tag_list = soup.select("tbody>tr")
```

```python
song_tag_list
```

```python
song_list = []

def 멜론차트_가져오기() :
    for song_tag in song_tag_list :
        song_id = song_tag["data-song-no"]
        song_rank = song_tag.select_one(".rank").text.strip()
        song_album = song_tag.select_one(".ellipsis.rank03 a").text.strip()
        song_title = song_tag.select_one(".ellipsis.rank01 a").text.strip()
        song_artist = song_tag.select_one(".ellipsis.rank02 a").text.strip()
        song_cover = song_tag.select_one("img")["src"]

        song_info = {
            "곡일렬번호" : song_id,
            "순위" : song_rank,
            "앨범" : song_album,
            "곡명" : song_title,
            "가수" : song_artist,
            "커버이미지_주소" : song_cover,
        }
        song_list.append(song_info)
        
    df = pd.DataFrame(song_list)
    df = df.set_index("곡일렬번호")
        
    return df
```

```python
df = 멜론차트_가져오기()

print(df.shape)
df.head()
```
