```python
import datetime
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36",
}

url = "https://www.melon.com/chart/index.htm"
res = requests.get(url, headers=headers)
res
```

```python
html = res.text
len(html)
```

```python
soup = BeautifulSoup(html, "html.parser")

press_wrap_tag_list = soup.select(".lst50")
len(press_wrap_tag_list)
```

```python
press_wrap_tag = press_wrap_tag_list[0]
press_name = press_wrap_tag.select_one(".lst50")

article_wrap_tag_list = press_wrap_tag.select(".lst50 td")

article_list = []

for article_wrap_tag in article_wrap_tag_list:
    article_rank = article_wrap_tag.select_one(".lst50 td t_center").text
    article_album = article_wrap_tag.select_one(".lst50 td rank03").text
    article_title = article_wrap_tag.select_one(".lst50 td rank01 title").text
    article_artist =article_wrap_tag.select_one(".lst50 td rank02").text
    article_cover_url = article_wrap_tag.select_one(".lst50 td wrap")["src"]
    article_songid = article_wrap_tag.select_one(".lst50")[""]

    article = {
        "press": press_name,
        "url": article_cover_url,
        "title": article_title,
        "rank": article_rank,
        "album": article_album,
        "artist": article_artist,
        "songID": article_songid
    }
    article_list.append(article)

article_list
```
