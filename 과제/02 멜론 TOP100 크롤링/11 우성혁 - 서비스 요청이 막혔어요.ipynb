```python
from bs4 import BeautifulSoup
import requests
from urllib.parse import urljoin
import pandas as pd
```

```python
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
}

url = "https://www.melon.com/chart/index.htm"
res = requests.get(url, headers = headers)
res
```

```python
html = res.text
soup = BeautifulSoup(html, "html.parser")

tag_list = soup.select(".lst50")
song_list = []

for tag in tag_list:
    song_no = tag["data-song-no"]
    song_rank = tag.select_one(".rank").text +"위"
    song_album = tag.select_one("a")["title"]
    song_title = tag.select_one("input")["title"]
    song_singer = tag.select_one(".checkEllipsis").text
    song_image = tag.select_one("img")["src"]
    
    song_list.append(
        {
            "no":song_no,
            "rank" : song_rank,
            "title":song_title.replace("곡 선택",""),
            "album":song_album,
            "singer":song_singer,
            "image":song_image
            
        })
    
df = pd.DataFrame(song_list)
df
```
