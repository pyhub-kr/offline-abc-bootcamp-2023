```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
headers = {
      'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
    }
```

```python
url ="https://www.melon.com/chart/index.htm"
res = requests.get(url,headers=headers)
res
```

```python
html = res.text
len(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
song_list =[]
```

```python
song_list =[]

"""
특성자 구문
[attr*=value]
attr이라는 특성값을 가지고 있으며, 값 안에 value라는 문자열이 적어도 하나 이상 존재한다면 이 요소를 선택합니다.


"""
for song_tag in soup.select('#tb_list tbody tr'):
    곡일렬번호 = song_tag['data-song-no']
    커버이미지_주소 = song_tag.select_one('img')['src']
    곡명 = song_tag.select_one('a[href*=playSong]').text
    가수 = song_tag.select_one('a[href*=goArtistDetail]').text
    앨범 = song_tag.select_one('a[href*=goAlbumDetail]')['title']
    순위 = song_tag.select_one('.rank').text
    
    song = {
            '곡일렬번호': 곡일렬번호,
            '커버이미지_주소': 커버이미지_주소,
            '곡명': 곡명,
            '가수': 가수,
            '앨범': 앨범,
            '순위': 순위,
            
        }
    song_list.append(song)

    print(song)
```

```python
song_list.append(song)
song_list
```

```python
# set_index : 기존의 행 인덱스를 제거하고 데이터 열 중 하나를 인덱스로 설정
pd.DataFrame(song_list,columns=['곡일렬번호','순위','앨범', '곡명', '가수', '커버이미지_주소']).set_index("곡일렬번호")
```
