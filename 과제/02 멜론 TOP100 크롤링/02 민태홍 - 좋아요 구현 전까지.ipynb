```python
num_tag_list=soup.select("tbody tr")
print(num_tag_list[0])
num_tag_list[0]["data-song-no"]
```

```python
# 반복문 시작
melon_chart= []
chart_tag_list = soup.select("tbody tr")  # len () = 100
c=0

for tag in chart_tag_list :
    c+=1
    print(c)
    if c>100 :        # 왜 100번 실행한 뒤 다시 처음부터 다시 시작할까?
        break   
    num_tag=soup.select_one("tbody tr")["data-song-no"]
    rank_tag = tag.select_one(".rank ").text
    album_tag = tag.select_one(".ellipsis.rank03 a").text.strip()
    song_tag = tag.select_one(".ellipsis").text.strip()
    singer_tag = tag.select_one(".ellipsis.rank02 a").text.strip()
    img_tag = tag.select_one(".wrap img")['src']
    
    chart = {
        "곡일련번호": num_tag,
        "순위": rank_tag,
        "앨범": album_tag,
        "곡명": song_tag,
        "가수": singer_tag,
        "커버이미지_주소": img_tag,
    }
    
    melon_chart.append(chart)
    
    
print(melon_chart)
```

```python
df =pd.DataFrame(melon_chart)
곡일련번호 = []
for i in df['곡일련번호']:
    곡일련번호.append(i)
df = df.set_index('곡일련번호')
df
```
