```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
}
```

```python
url = "https://www.melon.com/chart/index.htm"
res = requests.get(url, headers=headers)
res
```

```python
html = res.text
len(html)
```

```python
soup = BeautifulSoup(html, "html.parser")
list = soup.select('.service_list_song tr[data-song-no]')
list[0]
```

```python
top100_chart_list = []

for song in list:
    rank = song.select_one('.rank').text.strip()
    title = song.select_one('.ellipsis.rank01').text.strip()
    artist = song.select_one('.ellipsis.rank02').text.strip()
    album = song.select_one('.ellipsis.rank03').text.strip()
    img = song.select_one('img').get('src')
    
    top100_chart = {"순위":rank,
                    "곡명":title,
                    "아티스트":artist,
                    "앨범":album,
                    "대표사진 링크":img
                   }
    top100_chart_list.append(top100_chart)
    
    # print(f"순위: {rank}, 곡명: {title}, 아티스트: {artist}, 앨범: {album}, 대표사진 링크: {}")


top100_chart_list
```

```python
df = pd.DataFrame(top100_chart_list)
df.head()
```
