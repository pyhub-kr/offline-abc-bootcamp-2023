```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
}

url ="https://www.melon.com/chart/index.htm"
res=requests.get(url,headers=headers)
html=res.text
```

```python
soup = BeautifulSoup(html,"html.parser")
song_tag_list = soup.select("tbody>tr")
```

```python
song_list = []  # ADDED

for song_tag in song_tag_list:
    song_num = song_tag["data-song-no"]
    song_sco = song_tag.select_one(".rank").text.strip()
    song_album = song_tag.select_one(".ellipsis.rank03 a").text.strip()
    song_name = song_tag.select_one(".ellipsis.rank01 a").text.strip()
    song_singer = song_tag.select_one(".ellipsis.rank02 a").text.strip()
    song_image = song_tag.select_one("img")["src"]
    
    song_list.append({
            "곡일련번호": song_num,
            "순위": song_sco,
            "앨범": song_album,
            "곡명":song_name,
            "가수": song_singer,
            "커버이미지_주소":song_image,
        })


song_list
```

```python
df = pd.DataFrame(song_list)
print(df.shape)
df.head()
```
